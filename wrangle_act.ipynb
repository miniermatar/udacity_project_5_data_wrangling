```python
import pandas as pd
import requests 
import re
import os
from timeit import default_timer as timer
import tweepy
import json
import datetime
import numpy as np
import matplotlib.pyplot as plt
```

Reading `twitter-archive-enhanced.csv` file provided for the project

```python
df=pd.read_csv('twitter-archive-enhanced.csv')
df.head()
```

Downloading the data from the image prediction algorithm

```python
url="https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv"
```

```python
image_pred_file=requests.get(url)
with open(os.path.join(url.split('/')[-1]),mode='wb') as file:
        file.write(image_pred_file.content)
df_image=pd.read_csv(url.split('/')[-1], sep='\t')
df_image.head()
```

Collecting retweet count and favorite count from Twitter API

```python
#loading twitter API keys from file to avoid pushing the keys into github
keys=pd.read_csv('Keys.csv', names=['type','key'])
consumer_key = keys['key'][0]
consumer_secret = keys['key'][1]
access_token = keys['key'][2]
access_secret = keys['key'][3]

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
error=open('error.txt','w') #file to store the tweet IDs that were no found
jsondata=open('tweet_json.txt','w')
log=open('log.txt','w') # Log with the tweets stored in tweet_json.txt
for id in df.tweet_id:
    start = timer()
    try:
        tweet = api.get_status(id, tweet_mode='extended')
        json.dump(tweet._json,jsondata)
        jsondata.write('\n')
    except:
        error.write(str(id)+'\n')    
    end = timer()
    status='ID: {} - time: {}, current time: {}'.format(id,end-start,str(datetime.datetime.now())) 
    print(status)
    log.write(status+'\n')
log.close()
jsondata.close()
error.close()
```

```python
#Reading error file
errors=[]
error=open('error.txt','r')
for item in error:
    errors.append(item.strip())
error.close()
len(errors)
```

```python
#Extracting data from JSON
tweet_id=[]
retweet_count=[]
favorite_count=[]
with open("tweet_json.txt") as json_file:
    for line in json_file.readlines():
        data = dict(json.loads(line))
        tweet_id.append(data['id'])
        retweet_count.append(data['retweet_count'])
        favorite_count.append(data['favorite_count'])
        
json_dict={'tweet_id':tweet_id,'retweet_count':retweet_count,'favorite_count':favorite_count}
df_json=pd.DataFrame.from_dict(json_dict)
df_json.head()
```

### Visualization and Insigths

Let's first see the correlation between the rating score and the favorite counts

```python
#To ensure the plots are adjusted to the proper size
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
fig, ax1 = plt.subplots()
plt.tight_layout()
ax1.plot(df_clean['rating_score'], df_clean['favorite_count'] ,label='Favorite', c='b', ms=5, marker='o',ls=None, lw=0, markevery=1)
ax1.set(xlim=(0, 1.6))
ax1.set_xlabel('Rating score')
ax1.set_ylabel('Favorite count')
ax1.set_title('Relationship between score and favorite counts')
plt.savefig('rating_vs_favorite_counts.png', format='png', dpi=100);

```

As expected, higher favorite counts translated to higher score. Now let's see if there is a relationship between the favorite counts and the retweet counts

```python
fig, ax1 = plt.subplots()
ax1.plot(df_clean['retweet_count'], df_clean['favorite_count'] ,label='Favorite', c='b', ms=5, marker='o',ls=None, lw=0, markevery=1)
ax1.legend(loc=0)
ax1.set_ylabel('Favorites count')
ax1.set_xlabel('Retweet count')
ax1.set_title('Correlation between Retweets and Favorite counts');

```

There seems to be a linear correlation between retweets and favorites counts. Let's fit the data with a linear model and calculate the R-squared.

```python
import statsmodels.api as sm
df_clean['intercept']=1
lm=sm.OLS(df_clean['favorite_count'],df_clean[['intercept','retweet_count']])
modelfit=lm.fit()
print(modelfit.params)
modelfit.summary()
```

The model fitting shows a R-squared value of 0.859, which indicates a good correlation. Let's add the fitting to the previous plot.

```python
fig, ax1 = plt.subplots()
ax1.plot(df_clean['retweet_count'], df_clean['favorite_count'] ,label='Favorite', c='b', ms=5, marker='o',ls=None, lw=0, markevery=1)
#plotting model fitting
x = np.linspace(0,80000,10) # Generating the X axis values for the model fitting line
y = modelfit.params.retweet_count*x+modelfit.params.intercept #model equation based on OLS
ax1.plot(x,y, label='Model Linear Fit', c='r', ms=5, marker=None,ls="--", lw=2) # model fit plot
ax1.legend(loc=0)
ax1.set_ylabel('Favorites count')
ax1.set_xlabel('Retweet count')
ax1.set_title('Correlation between Retweets and Favorite counts')
plt.savefig('retweet_vs_favorites.png', format='png', dpi=100);

```
