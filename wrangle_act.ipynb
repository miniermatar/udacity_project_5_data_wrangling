Downloading the image prediction data

````````````````````````````````````python
url="https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv"
````````````````````````````````````

````````````````````````````````````python
image_pred_file=requests.get(url)
with open(os.path.join(url.split('/')[-1]),mode='wb') as file:
        file.write(image_pred_file.content)
df_image=pd.read_csv(url.split('/')[-1], sep='\t')
df_image.head()
````````````````````````````````````

Collecting retweet count and favorite count from Twitter APi

````````````````````````````````````python
#loading twitter API keys
keys=pd.read_csv('Keys.csv', names=['type','key'])
consumer_key = keys['key'][0]
consumer_secret = keys['key'][1]
access_token = keys['key'][2]
access_secret = keys['key'][3]

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
````````````````````````````````````

````````````````````````````````````python
error=open('error.txt','w') #file to store the tweet IDs that were no found
jsondata=open('tweet_json.txt','w')
log=open('log.txt','w')
for id in df.tweet_id:
    start = timer()
    try:
        tweet = api.get_status(id, tweet_mode='extended')
        json.dump(tweet._json,jsondata)
        jsondata.write('\n')
    except:
        error.write(str(error)+'\n')    
    end = timer()
    status='ID: {} - time: {}, current time: {}'.format(id,end-start,str(datetime.datetime.now())) 
    print(status)
    log.write(status+'\n')
log.close()
jsondata.close()
error.close()
````````````````````````````````````

````````````````````````````````````python
#Reading error file
errors=[]
error=open('error.txt','r')
for item in error:
    errors.append(item.strip())
error.close()
````````````````````````````````````

````````````````````````````````````python
len(errors)
````````````````````````````````````

````````````````````````````````````python
#Extracting data from JSON
tweet_id=[]
retweet_count=[]
favorite_count=[]
with open("tweet_json.txt") as json_file:
    for line in json_file.readlines():
        data = dict(json.loads(line))
        tweet_id.append(data['id'])
        retweet_count.append(data['retweet_count'])
        favorite_count.append(data['favorite_count'])
        
json_dict={'tweet_id':tweet_id,'retweet_count':retweet_count,'favorite_count':favorite_count}
df_json=pd.DataFrame.from_dict(json_dict)
df_json.head()
````````````````````````````````````
